```python
import pandas as pd
import numpy as np
from scipy.stats import f

df = pd.read_csv("top2022.csv")

nilai = df["Nilai.Total"]
provinsi = df["Provinsi"]

groups = df.groupby("Provinsi")["Nilai.Total"]

k = groups.ngroups
N = len(df)
overall_mean = nilai.mean()

SSB = 0
SSW = 0

for prov, data in groups:
    ni = len(data)
    group_mean = data.mean()

    SSB += ni * (group_mean - overall_mean)**2

    SSW += ((data - group_mean)**2).sum()

SST = SSB + SSW

df_between = k - 1
df_within = N - k
df_total = N - 1

MSB = SSB / df_between
MSW = SSW / df_within

F_value = MSB / MSW

p_value = 1 - f.cdf(F_value, df_between, df_within)

print("===== HASIL ANOVA SATU ARAH (MANUAL) =====")
print(f"Total Provinsi      : {k}")
print(f"Total Data          : {N}")
print("---------------------------------------------")
print(f"SSB (Between)       : {SSB}")
print(f"SSW (Within)        : {SSW}")
print(f"SST (Total)         : {SST}")
print("---------------------------------------------")
print(f"df Between          : {df_between}")
print(f"df Within           : {df_within}")
print(f"df Total            : {df_total}")
print("---------------------------------------------")
print(f"MSB                 : {MSB}")
print(f"MSW                 : {MSW}")
print("---------------------------------------------")
print(f"F-hitung            : {F_value}")
print(f"p-value             : {p_value}")
print("=============================================")

anova_table = pd.DataFrame({
    "SS": [SSB, SSW, SST],
    "df": [df_between, df_within, df_total],
    "MS": [MSB, MSW, ""],
    "F": [F_value, "", ""]
}, index=["Between (Provinsi)", "Within (Error)", "Total"])

print("\n===== TABEL ANOVA =====")
print(anova_table)
```
