```python
import csv
import math

def mean(data):
    return sum(data) / len(data)

def std_dev(data):
    mu = mean(data)
    variance = sum((x - mu) ** 2 for x in data) / len(data)
    return math.sqrt(variance)

def fungsi_normal(mu, sigma, x):
    return (1 / (sigma * math.sqrt(2 * math.pi))) * math.exp(-0.5 * ((x - mu) / sigma) ** 2)

def peluang_normal(mu, sigma, a, b, n=1000):
    h = (b - a) / n
    total = 0
    for i in range(n + 1):
        x = a + i * h
        fx = fungsi_normal(mu, sigma, x)
        if i == 0 or i == n:
            total += fx
        else:
            total += 2 * fx
    return (h / 2) * total

data = []
with open('top2022.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(float(row['Nilai.Total']))

mu = mean(data)
sigma = std_dev(data)

print(f"Rata-rata (μ): {mu:.4f}")
print(f"Simpangan baku (σ): {sigma:.4f}")

a = mu - sigma
b = mu + sigma
p = peluang_normal(mu, sigma, a, b)
print(f"Peluang {a:.2f} < X < {b:.2f} = {p:.6f}")
```
